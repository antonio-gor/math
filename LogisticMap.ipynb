# Logistic Map Bifurcation Diagram

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Main Functions

```python
def logistic_fn(r, p=0.5, n=1000):
    """ Return N logistic values for given growth rate R and initial population P. """
    x = p + np.zeros(n) # Initial population x[0] is P
    for i in np.arange(1, n):
        x[i] = r * x[i-1] * (1 - x[i-1])
    return x

def logistic_map(p=0.5, num=2000, rmin=1, rmax=4):
    """ Returns the limits for varying growth rates. """
    all_limits = []
    r_vals = np.linspace(rmin, rmax, num)
    for r in r_vals: # Apply logistic fn for each growth rate
        logistic_vals = logistic_fn(r, p)
        last_ten_percent_vals = logistic_vals[-int(len(logistic_vals)*0.1):]
        limits = np.unique(last_ten_percent_vals)
        all_limits.append(limits)
    return r_vals, all_limits

def bifurcation_diagram(p=0.5, num=2000, rmin=1, rmax=4, ymin=0, ymax=1):
    """ Plots the bifurcation diagram for the logistic map. """
    r_vals, all_limits = logistic_map(p, num, rmin, rmax)
    plt.figure(figsize=(12, 8))
    for r, x in zip(r_vals, all_limits):
        plt.scatter([r]*len(x), x, s=0.1, c='k')
    plt.title("Bifurcation Diagram for the Logistic Map")
    plt.xlabel("Growth Rate (r)")
    plt.ylabel("Population (x)")
    plt.ylim(ymin, ymax)
```
